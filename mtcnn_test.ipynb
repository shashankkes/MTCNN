```
# face detection with mtcnn on a photograph
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
import cv2 

# draw an image with detected objects
def draw_image_with_boxes(filename, result_list):
	# load the image
	data = pyplot.imread(filename)
	# plot the image
	pyplot.imshow(data)
	# get the context for drawing boxes
	ax = pyplot.gca()
	# plot each box
	for result in result_list:
		# get coordinates
		x, y, width, height = result['box']
		# create the shape
		rect = Rectangle((x, y), width, height, fill=False, color='red')
		# draw the box
		ax.add_patch(rect)
		# draw the dots
		for key, value in result['keypoints'].items():
			# create and draw dot
			dot = Circle(value, radius=2, color='red')
			ax.add_patch(dot)
	# show the plot
	pyplot.show()
filename = '/content/drive/MyDrive/_+91 98078 90581_ 20161229_135628.jpg'
# load image from file
pixels = pyplot.imread(filename)
# create the detector, using default weights
detector = MTCNN()
# detect faces in the image
faces = detector.detect_faces(pixels)
# display faces on the original image
draw_image_with_boxes(filename, faces)

```

```
import cv2
from mtcnn.mtcnn import MTCNN
det=MTCNN()
#to collect the bounding box list
bounding_box=[]
frame_num=[] #this store which frame contain face bounding box
for i in range(0,144-1):
  img=cv2.imread("/content/drive/MyDrive/Test_folder/frame_no{}.jpg".format(i))
  result=det.detect_faces(img)
  #if the retured result is not empty then it will save 
  if(len(result)!=0):
    cv2.imwrite("/content/drive/MyDrive/SK_face/frame_num{}.jpg".format(i),img)
    #append the frame num 
    frame_num.append(i)
    #append the frame face box list from the result 
    bounding_box.append(result[0]['box'])
  
```

This is for extracting the facial co-ordinates

```
from mtcnn import MTCNN
import cv2
det=MTCNN()
bounding_box=[]
frame_num=[]
for i in range(vlen):
  img=cv2.imread("/content/drive/MyDrive/Extracted_faces/PK/PK_frame{}.jpg".format(i))
  result=det.detect_faces(img)
  if len(result)!=0:
    bounding_box.append(result[0]['box'])
    frame_num.append(i)
```

This is cropping the face

```
import cv2
count=0
for i in bounding_box:
  img=cv2.imread("/content/drive/MyDrive/Extracted_faces/PK/PK_frame0.jpg".format(frame_num[count]))
  x=i[0]
  y=i[1]
  x_h=i[2]+x
  y_h=i[3]+y
  if y<0:
    crop_img=img[(-1*y):y_h,x:x_h]
  else:
    crop_img=img[y:y_h,x:x_h]
  cv2.imwrite("/content/drive/MyDrive/Cropped_faces/PK_cropped_face/PK_cropped_face{}.jpg".format(frame_num[count]),crop_img)
  count+=1

```
